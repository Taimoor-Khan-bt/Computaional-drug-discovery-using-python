```python
import pandas as pd
```

```python
df = pd.read_csv("bioactivity_preprocessed_data.csv")
```

```python
# Generate Lipinski descriptors
# Import the Neccessary libraries

import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
def linpinski(smiles, verbose=False):
    
    moldata = list()
    for elem in smiles:
        mol = Chem.MolFromSmiles(elem)
        moldata.append(mol)
        
    basedata = np.arange(1,1)
    i = 0
    
    for mol in moldata:
        desc_molwt = Descriptors.MolWt(mol)
        desc_molLogp= Descriptors.MolLogP(mol)
        desc_numHdonors = Descriptors.NumHDonors(mol)
        desc_numHacceptors = Descriptors.NumHAcceptors(mol)
        
        row = np.array([desc_molwt,
                   desc_molLogp,
                   desc_numHdonors,
                   desc_numHacceptors])
        if i == 0:
            basedata=row
        else:
            basedata= np.vstack([basedata, row])
        i += 1
    column_names = ["MW", "MolLogP", "NumHDonors", "NumHAcceptors"]
    descriptors = pd.DataFrame(data=basedata, columns = column_names)
    
    return descriptors
```

```python
df_lipinski = linpinski(df.canonical_smiles)
```
